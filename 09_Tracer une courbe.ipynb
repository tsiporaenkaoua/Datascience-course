```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
df.price_paid = df.price_paid.apply(lambda x: x.replace("$",""))#df.price_paid peut aussi etre représenté de cette maniere : df["price_paid"]
df.price_paid =df.price_paid.astype(float) # changer le type d'une donnée
```

```python
df.groupby("date")["price_paid"].sum().plot(figsize=[20,10])#plot qui gere les graphique grace a matplotlib qui na pas besoin detre importé (mais ils a besoin d'être installé, ici sur jupiter lab il est deja installé)
```

```python
df.groupby("gender")["price_paid"].sum().plot.pie(legend=True)
```

```python
df.groupby("country")["price_paid"].sum().plot.bar(rot=45, legend=True)
```
